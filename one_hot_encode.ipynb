# One Hot Encoding specific columns

```python
import pandas as pd

df = pd.read_csv("data/acled.csv")
df.head()
```

```python
ohe_df = pd.get_dummies(df, columns=["sub_event_type", "inter1", "region"])

ohe_df.astype(int)
ohe_df.head()
```

Also convert `event_time` column into Datetime objects

```python
from datetime import datetime

ohe_df["event_date"] = pd.to_datetime(ohe_df["event_date"])

assert isinstance(ohe_df["event_date"].iloc[0], datetime)
```

We also notice that our one hot encoded columns seems to be booleans of True and False. Let's convert these to 0s and 1s

```python
# Iterate through each column
for col in ohe_df.columns:
    if ohe_df[col].dtype == 'bool':  # Check if the column has boolean values
        ohe_df[col] = ohe_df[col].astype(int)  # Convert True/False to 1/0
```

```python
ohe_df.to_csv("data/ohe_acled.csv")
```
